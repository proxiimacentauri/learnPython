## Advance Operations on Lists

A list can hold scalars, but it can also hold other lists as elements of a list.
<br>



### List Comprehension

For example if we want to represent a chessboard's starting position in which the 2nd row is filled with 8 WHITE PAWNS.

`[expression for element in list if conditional]`

is equivalent of: 

`for element in list:
    if conditional:
        expression`





```python
WHITE_PAWN=1 # Assume that WHITE_PAWN is a pre-def symbol.
row=[]

for i in range(8):
    row.append(WHITE_PAWN)
    
### This below is an easier representation of the above code    ####
    
row1 = [WHITE_PAWN for i in range(8)]
```

Similar Examples are:

```python
squares = [x ** 2 for x in range(10)]
print(squares)

twos = [2 ** i for i in range(8)]
print(twos)

odds = [x for x in squares if x % 2 != 0 ]
print(odds)
```

### 3-D Arrays

Imagine a hotel. It's a huge hotel consisting of 3 buildings, 15 floors each and 20 rooms on each floor. For this, you need an array which can collect and process information on the occupied/free rooms.

Here,

* 1st Index = 0 through 2 selectes the Buildings
* 2nd Index = 0 through 15 selects the Floor
* 3rd Index = 0 through 20 selects the Room

`cube = [[[':(', 'x', 'x'],
         [':)', 'x', 'x'],
         [':(', 'x', 'x']],`

        `[[':)', 'x', 'x'],
         [':(', 'x', 'x'],
         [':)', 'x', 'x']],`

       ` [[':(', 'x', 'x'],
         [':)', 'x', 'x'],
         [':)', 'x', 'x']]]`




```python
rooms=[[[False for i in range(20)] for j in range(15)] for k in range(3)]
rooms[1][9][13] = True
#rooms[3][15][20] = True # This will be out of range
rooms[2][14][19] = True
```

```python
occupied=0
for i in rooms: # i = will have a 
    #print(i)
    for j in i:
        #print(j)
        for k in j:
            if k == True:
                occupied+=1
print(occupied)
```
