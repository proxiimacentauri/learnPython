## Packages

While a module is designed to couple together some related entities (functions, variables, constants, etc.), a package is a container which enables the coupling of several related modules under one common name. Such a container can be distributed as-is (as a batch of files deployed in a directory sub-tree) or it can be packed inside a zip file.

* Imagine that in the not-so-distant future you and your associates write a large number of Python functions.

* Your team decides to group the functions in separate modules, and this is the final result of the ordering:

```python
#! /usr/bin/env python3

""" module: alpha """

def funA():
    return "Alpha"

if __name__ == "__main__":
    print("I prefer to be a module.")

```

### Initialization (Module Tree ---> Package) (\_\_init\_\_.py)

The contents of the file `__init__.py` is executed when any of the package modules is **imported**.<br>

If you don't want any special initializations, you can leave the file empty, **but you mustn't omit it.**
<br>

The location of `__init__.py` can be anywhere in the package (root, or in any-sub-folder).
<br><br>

A Python file named __init__.py is implicitly run when a package containing it is subject to import, and is used to initialize a package and/or its sub-packages (if any). The file may be empty, but must not be absent.
