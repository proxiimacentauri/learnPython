## Tic-Tac-Toe

Program which pretends to play tic-tac-toe with the user

### Rules

* Computer = `X`
* User = `O`
* First move of Computer and it will always put a `X` in middle of board.
* All the squares are numbered as below, and user inputs into which square they want to place `O`. (it should be valid >0 and <10 and and int and cant point to already occupied square)
* There are 4 possible verdicts:
    + Continue
    + End with Tie
    + Ends with Computer Win
    + Ends with User Win
* Computer responds with its move check is repeated.
* A random choice implementation is enough.

### Requirements

Implement the following features:

* Board should be stored as a 3 element list, while each element is another 3 element list. (The inner list represents ROWS) Syntax: board[row][column]
* Each of inner lists elements can contain `O` or `X` or a digit representing square's number.
* Boards presence should be exactly the same as below. 
* Drawing a random integer `randrange()`

```python
from random import randrange

for i in range(10): # i is a counter
    print(randrange(4))  # It means it will pick from 0,1,2,3 and output them randomly  
```

```python
def display_board(board):
    # The function accepts one parameter containing the board's current status
    # and prints it out to the console.
    
    p,q,r=board[0][0],board[0][1],board[0][2]
    s,t,u=board[1][0],board[1][1],board[1][2]
    v,w,y=board[2][0],board[2][1],board[2][2]

    print(
    '''
    +-------+-------+-------+
    |       |       |       |
    |   {}   |   {}   |   {}   |
    |       |       |       |
    +-------+-------+-------+
    |       |       |       |
    |   {}   |   {}   |   {}   |
    |       |       |       |
    +-------+-------+-------+
    |       |       |       |
    |   {}   |   {}   |   {}   |
    |       |       |       |
    +-------+-------+-------+
    '''.format(p,q,r,s,t,u,v,w,y))

def enter_move(board):
    # The function accepts the board current status, asks the user about their move, 
    # checks the input and updates the board according to the user's decision.
    user=int(input("Enter your move: "))
    if user < 0 and user >=10:
            print('Invalid box: ')
            user=int(input('Please a number from 1 to 9: '))

    if user == 1:
        if board[0][0]=='X' or board[0][0]=='O':
           print('Box is Occupied: ')
           user=int(input('Please a number from 1 to 9: ')) 
        board[0][0]='O'
    if user == 2:
        if board[0][1]=='X' or board[0][1]=='O':
           print('Box is Occupied: ')
           user=int(input('Please a number from 1 to 9: ')) 
        board[0][1]='O'
    if user == 3:
        if board[0][2]=='X' or board[0][2]=='O':
           print('Box is Occupied: ')
           user=int(input('Please a number from 1 to 9: ')) 
        board[0][2]='O'
    if user == 4:
        if board[1][0]=='X' or board[1][0]=='O':
           print('Box is Occupied: ')
           user=int(input('Please a number from 1 to 9: ')) 
        board[1][0]='O'
    if user == 5:
        if board[1][1]=='X' or board[1][1]=='O':
           print('Box is Occupied: ')
           user=int(input('Please a number from 1 to 9: ')) 
        board[1][1]='O'
    if user == 6:
        if board[1][2]=='X' or board[1][2]=='O':
           print('Box is Occupied: ')
           user=int(input('Please a number from 1 to 9: ')) 
        board[1][2]='O'
    if user == 7:
        if board[2][0]=='X' or board[2][0]=='O':
           print('Box is Occupied: ')
           user=int(input('Please a number from 1 to 9: ')) 
        board[2][0]='O'
    if user == 8:
        if board[2][1]=='X' or board[2][1]=='O':
           print('Box is Occupied: ')
           user=int(input('Please a number from 1 to 9: ')) 
        board[2][1]='O'
    if user == 9:
        if board[2][2]=='X' or board[2][2]=='O':
           print('Box is Occupied: ')
           user=int(input('Please a number from 1 to 9: ')) 
        board[2][2]='O'
    
    display_board(board)
    draw_move(board)
            
def make_list_of_free_fields(board):
    # The function browses the board and builds a list of all the free squares; 
    # the list consists of tuples, while each tuple is a pair of row and column numbers.
    free_spots=[]
    for i in range(3):
        for j in range(3):
               if board[i][j] != 'X' and board[i][j] != 'O':
                    #print('[',i,']','[',j,']'," Is Free")
                    free_spots.append((i,j))
    return free_spots
def draw_move(board):
    # The function draws the computer's move and updates the board.
    
    free_spots_to_draw=make_list_of_free_fields(board) #This is a list of tuples
    
    spot_to_draw=free_spots_to_draw.pop(0)
    
    for i in range(3):
        for j in range(3):
            if spot_to_draw == (i,j):
                board[i][j]='X'
    
    display_board(board)
def victory_for(board, sign):
    # The function analyzes the board status in order to check if 
    # the player using 'O's or 'X's has won the game
    if board[0][0]==sign and board[1][1]==sign and board[2][2]==sign:
        return True
    
    if board[0][2]==sign and board[1][1]==sign and board[2][0]==sign:
        return True
    
    
    if board[0][0]==sign and board[0][1]==sign and board[0][2]==sign:
        return True
    
    if board[1][0]==sign and board[1][1]==sign and board[1][2]==sign:
        return True
    
    if board[2][0]==sign and board[2][1]==sign and board[2][2]==sign:
        return True
    
    
    
    if board[0][0]==sign and board[1][0]==sign and board[2][0]==sign:
        return True
    
    if board[0][1]==sign and board[1][1]==sign and board[2][1]==sign:
        return True
    
    if board[0][2]==sign and board[1][2]==sign and board[2][2]==sign:
        return True
    
    return False
    
    
global board
board=[[1,2,3],
        [4,'X',6],
        [7,8,9]]

display_board(board)
for i in range(10):
    enter_move(board)
    if victory_for(board,'X'):
        print('Computer has Won')
        break
    if victory_for(board,'O'):
        print('User has Won')
        break
```
