## Foundations of OOP

* **Procedural Style**
    + It was a dominant approach and is still used today. 
    + It works very well for certian types of preojects.
    + it's possible to distinguish two different and completely separate worlds: the world of data, and the world of code. 
<br><br>
* **Object Approach**
    + The data and the code are enclosed together in the same world, divided into classes.
    + There is no clear border between data and code: they live as one in objects.
    + Every `class` is like a recipe which can be used when you want to create a useful object 
    + Every `object` has a set of traits (they are called properties or attributes - we'll use both words synonymously) and is able to perform a set of activities (which are called methods).
    + `Objects are incarnations` of ideas expressed in classes, like a cheesecake on your plate is an incarnation of the idea expressed in a recipe printed in an old cookbook.

#### Class


 A class is an idea (more or less abstract) which can be used to create a number of incarnations â€“ such an incarnation is called an object.

#### Inheritance


When a class is derived from another class, their relation is named inheritance.

<br>
The class which derives from the other class is named a subclass. The second side of this relation is named superclass. A way to present such a relation is an inheritance diagram.


#### Objects 


Objects are equipped with:<br>

* a name which identifies them and allows us to distinguish between them;
* a set of properties (the set can be empty)
* a set of methods (can be empty, too)

### What is an object?

A class (among other definitions) is a set of objects. **An object is a being belonging to a class.**
<br><br>
**An object is an incarnation of the requirements, traits, and qualities assigned to a specific class. This may sound simple, but note the following important circumstances. Classes form a hierarchy.**

<br>

Ex: Any personal car is an object belonging to the wheeled vehicles class. It also means that the same car belongs to all superclasses of its home class; therefore, it is a member of the vehicles class, too.

<br>

* Each subclass is more specialized (or more specific) than its superclass. Conversely, each superclass is more general (more abstract) than any of its subclasses.
* We've **presumed that a class may only have one superclass - this is not always true**

### What does an object have?

The object programming convention assumes that every existing object may be equipped with three groups of attributes:

<br>

* Name:  that uniquely identifies it within its home namespace
* Set of Indivisual Properties: which make it original, unique or outstanding (although it's possible that some objects may have no properties at all)
* Set of abilities to perform specific activities: able to change the object itself, or some of the other objects.

### Inheritance


Any object bound to a specific level of a class hierarchy inherits all the traits (as well as the requirements and qualities) defined inside any of the superclasses.

### Objects and Classess in Python


The class you define has nothing to do with the object: the existence of a class does not mean that any of the compatible objects will automatically be created. The class itself isn't able to create an object - you have to create it yourself, and Python allows you to do this.

<br>

The `pass` keyword fills the class with nothing. It doesn't contain any methods or properties.

```python
class TheSimplestClass:
    pass
```

**The act of creating an object of the selected class is also called an `instantiation` (as the object becomes an instance of the class).**
